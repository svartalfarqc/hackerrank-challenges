```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
import sys

# training array
train_array=[]
train_label=[]

# get n = number of training entries and m = number of features
n,m=[int(e) for e in input().split()]

for i in range(n):
    a=input().split()

    # remove actual string, we don't need it
    a.pop(0)

    # take label (+1 = good, -1 = bad)
    train_label.append(int(a[0]))
    a.pop(0)

    # get features
    train_array.append(
            {
                # get each pair of feature id (1 to m):value

                # get id:value pair for each space-separated value in the input string
                int(e[0]):float(e[1]) for e in [x.split(':') for x in a]
            }
        )


# DictVectorizer transforms lists of feature-value mappings to vectors.
# https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.DictVectorizer.html

from sklearn.feature_extraction import DictVectorizer
dict_vectorizer = DictVectorizer()

train = dict_vectorizer.fit_transform(train_array).toarray()

# Use Random Forest Classifier
# (read somewhere that SVM doesn't work)

from sklearn.ensemble import RandomForestClassifier

forest_classifier = RandomForestClassifier()
forest_classifier.fit(train,train_label)

n_tests = int(input())
test_string = []
test_array = []

for i in range(n_tests):
    a = input().split()

    # get test string
    test_string.append(a[0])
    a.pop(0)

    # get test parameters
    test_array.append({float(e[0]):float(e[1]) for e in [x.split(":") for x in a]})

test = dict_vectorizer.transform(test_array).toarray()
test_label = forest_classifier.predict(test)

for string,label in zip(test_string,test_label):
    print(string + ' %+d' % (label,) )
```
