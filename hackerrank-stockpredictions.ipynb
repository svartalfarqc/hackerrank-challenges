```python
# Solution based on this solution:
# https://datumguy.wordpress.com/2017/06/27/hackerranks-stocks-predictions-exercise-review/
# In fact it's pretty much the same for now, but I play to play with it someday
import numpy as np
 
alpha = 2 / 6
exp = np.array([alpha * (1 - alpha) ** (5 - i) for i in range(1, 6)]) + (1 - alpha) ** 5 / 5
 
def printTransactions(m, k, d, name, owned, prices):
    max_f = None
    min_f = None
 
    for i in range(k):
        
        
        # f = exponential moving average - mean
        f = getCEMA(prices[i]) - np.mean(prices[i])
        
        # There are many strategies, as clouds in the sky, but we’ll opt for a simple one:

        # Sell all stocks we have for the one with the highest spike.
        # Buy all stocks we can for the one with the hardest fall – and hope its price increases in the next days.
        
        # Highest spike
        if owned[i] > 0 and f > 0 and (max_f is None or f > max_f[0]):
            max_f = (f, '{0} SELL {1}'.format(name[i], owned[i]))
        # Lowest drop
        elif m > prices[i][-1] and d > 0 and f > 0 and (min_f is None or f < min_f[0]):
            min_f = (f, '{0} BUY {1}'.format(name[i], int(m / prices[i][-1])))
 
    # Print transactions of the day
    if max_f is not None and min_f is not None:
        print(2)
    elif max_f is not None or min_f is not None:
        print(1)
    else:
        print(0)
 
    if max_f is not None:
        print(max_f[1])
    if min_f is not None:
        print(min_f[1])
 
def getCEMA(a):
    """
    Get Customized Exponential Moving Average (CEMA)
    Since we’ll work with prices in windows of five days, EMA as it is won’t be useful because it’s a cumulative function, i.e. its calculation depends on all historic values. We customize EMA to distribute exponentially the values in the window.
    The equation is valid because it still is exponential and its distribution sums 1.
    """
    return np.dot(exp, a)
 
def readInputAndRun():
    line = input().split()
    m, k, d = float(line[0]), int(line[1]), int(line[2])
    name = []
    owned = []
    prices = []
    for i in range(k):
        line = input().split()
        name.append(line[0])
        owned.append(int(line[1]))
        prices.append(list(map(float, line[2:])))
    printTransactions(m, k, d, name, owned, prices)
 
readInputAndRun()
```
